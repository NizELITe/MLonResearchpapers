```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score


data = pd.read_csv("/content/lab7_task1_Final.csv")
data= data.fillna(data.mean())

desired_num_features = 10


data_divisions = [0.5, 0.7, 0.8]

results_df = pd.DataFrame(columns=['Data Division', 'Selected Features', 'Model Accuracy'])

for data_ratio in data_divisions:

    X = data.drop("target", axis=1)
    y = data["target"]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1 - data_ratio, random_state=42)


    selected_features = []

    while len(selected_features) < desired_num_features:

        svm_classifier = SVC(kernel='linear')
        svm_classifier.fit(X_train, y_train)


        feature_importance = abs(svm_classifier.coef_[0])


        min_importance_feature = X_train.columns[np.argmin(feature_importance)]
        selected_features.append(min_importance_feature)
        X_train = X_train.drop(min_importance_feature, axis=1)
        X_test = X_test.drop(min_importance_feature, axis=1)


        y_pred = svm_classifier.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)


    results_df = results_df.append({'Data Division': f'{data_ratio * 100}%-Split',
                                    'Selected Features': ', '.join(selected_features),
                                    'Model Accuracy': accuracy}, ignore_index=True)

print(results_df)
```
